```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.stock_img()
ax.coastlines()
ax.scatter(homicidios['lon'], homicidios['lat'], color='red', transform=ccrs.PlateCarree())
ax.set_extent([-130, -60, 20, 60], crs=ccrs.PlateCarree())
ax.set_title('homicidios en 55 ciudades por coordenadas')
plt.show()
```

```python
closed_no_arrest = homicidios[homicidios['disposition'] == 'Closed without arrest']
```

```python
grouped = closed_no_arrest.groupby('city').count()
grouped['percentage_closed_no_arrest'] = grouped['disposition'] / grouped['disposition'].sum() * 100
```

```python
print(grouped['percentage_closed_no_arrest'])  # porcentaje de casos cerrados sin arresto
```

```python
open_no_arrest = homicidios[homicidios['disposition'] == 'Open/No arrest']
```

```python
grouped = open_no_arrest.groupby('city').count()
grouped['percentage_open_no_arrest'] = grouped['disposition'] / grouped['disposition'].sum() * 100
```

```python
print(grouped['percentage_open_no_arrest'])  # porcentaje de casos abiertos sin arresto
```

```python
homicidios['victim_age'] = pd.to_numeric(homicidios['victim_age'], errors='coerce')
```

```python
media_edad_victimas=homicidios.groupby('city')['victim_age'].mean()
```

```python
homicidios.groupby('city')['victim_age'].mean()
```

```python
raza_mas_frecuente = homicidios.groupby(['city', 'victim_race'])['victim_race'].count().groupby(level=0).apply(lambda x: x.idxmax()[1])
```

```python
resultados = pd.concat([media_edad_victimas, raza_mas_frecuente], axis=1)
resultados.columns = ['media_edad_victimas', 'raza_mas_frecuente']
resultados
```
